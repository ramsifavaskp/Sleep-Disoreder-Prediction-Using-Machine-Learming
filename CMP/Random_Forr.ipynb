```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
```

```python
df.head()
```

```python
clean_df = df.drop(columns='Person ID', axis=1)
```

```python
Categorical_Col = []
Numerical_Col = []
```

```python
for col in clean_df.columns:
    if clean_df[col].dtype == 'object':
        Categorical_Col.append(col)

    else:
        Numerical_Col.append(col)
```

```python
Categorical_Col
```

```python
Numerical_Col
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
```

```python
for col in Categorical_Col:
    clean_df[col] = LE.fit_transform(clean_df[col])
```

```python
from sklearn import preprocessing
#make an instance of Label Encoder
label_encoder = preprocessing.LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])
df['Occupation'] = label_encoder.fit_transform(df['Occupation'])
df['BMI Category'] = label_encoder.fit_transform(df['BMI Category'])
df['Sleep Disorder'] = label_encoder.fit_transform(df['Sleep Disorder'])
df['Blood Pressure'] = label_encoder.fit_transform(df['Blood Pressure'])
df['Sleep Disorder'] = label_encoder.fit_transform(df['Sleep Disorder'])
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = clean_df.iloc[:,:-1]
X
```

```python
y = clean_df['Sleep Disorder']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RNF = RandomForestClassifier()
```

```python
RNF.fit(X_train, y_train)
```

```python
y_hat = RNF.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print(accuracy_score(y_test, y_hat))
```

```python
print(classification_report(y_test, y_hat))
```
